```python
import sagemaker
from sagemaker import get_execution_role

```

```python
sagemaker_session = sagemaker.Session()
role = get_execution_role()
```

```python
from sagemaker import image_uris

image_uri = image_uris.retrieve(
    framework='xgboost', 
    region='ap-south-1',  # use your region
    version='1.5-1'       # change version if needed
)
```

```python
from sagemaker.model import Model

model = Model(
    model_data='s3://soil-prediction-bucket/model/model.tar.gz',
    role=role,
    entry_point='inference.py',
    image_uri=image_uri,
    sagemaker_session=sagemaker_session
)

```

```python
import joblib
joblib.dump(model, 'model/soil_model.pkl')
```

```python
predictor1 = model.deploy(
    initial_instance_count=1,
    instance_type='ml.t2.medium',
    endpoint_name='soil-quality-endpoint-v6'
)
```

```python
# Example input: [N, P, K, temp, humidity, pH, rainfall]
response = predictor1.predict([90, 42, 43, 23.5, 80, 6.5, 120])
print("Predicted soil type:", response)
```
